 # Split the X&Y

```python
X1 = df.drop("Channel",axis=1)
Y1 = df["Channel"]
```

```python
Y1[Y1 == 2] = 0
Y1[Y1 == 1] = 1
```

 # Train_Test_Split

```python
from sklearn.model_selection import train_test_split
X1_train,X1_test,Y1_train,Y1_test = train_test_split(X1,Y1,test_size=0.1,random_state = 21)
X1_train
```

 # Importing XGBoost Classifier

```python
from xgboost import XGBClassifier
params = {
            'objective' : 'binary:logistic',
            'max_depth' : 4,
            'alpha'     : 10,
            'learning_rate' : 1.0,
            'n_estimators'  : 100,
}
xgb_clfn1 = XGBClassifier(**params)
xgb_clfn1.fit(X1_train,Y1_train)
```

# Prediction from our Model

```python
Y1_pred = xgb_clfn1.predict(X1_test)
```

```python
snd.jointplot(Y1_test)
```

 # Checking Accuracy

```python
from sklearn.metrics import accuracy_score
print("XGBoost model accuracy score: {0:0.4f}".format(accuracy_score(Y1_test,Y1_pred)))
```
