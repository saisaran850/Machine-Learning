```python
# Defining x
X = df[["Annual Income (k$)","Spending Score (1-100)"]]
```

```python
# kmeans
k_means = KMeans()
k_means.fit(X)
```

```python
# Predicting x
k_means = KMeans()
k_means.fit_predict(X)
```

```python
# Using Elbow method to find no.of clusters
# WCSS Means With In Clusters Sum Of Squares
wcss = []
for i in range(1,11):
    k_means = KMeans(n_clusters = i)
    k_means.fit(X)
    wcss.append(k_means.inertia_)
```

```python
wcss
```

```python
# Using matplotlib
import matplotlib.pyplot as pse
pse.plot(range(1,11),wcss)
pse.title("Elbow Method")
pse.xlabel("Number of Clusters")
pse.ylabel("wcss")
pse.show()
```

```python
# Model Training
X = df[["Annual Income (k$)","Spending Score (1-100)"]]
k_means = KMeans(n_clusters=5,random_state=42)
y_means = k_means.fit_predict(X)
y_means
```

```python
pse.scatter(X.iloc[y_means==0,0],X.iloc[y_means==0,1],s=100,c="pink",label="Cluster 1")
pse.scatter(X.iloc[y_means==1,0],X.iloc[y_means==1,1],s=100,c="yellow",label="Cluster 2")
pse.scatter(X.iloc[y_means==2,0],X.iloc[y_means==2,1],s=100,c="blue",label="Cluster 3")
pse.scatter(X.iloc[y_means==3,0],X.iloc[y_means==3,1],s=100,c="green",label="Cluster 4")
pse.scatter(X.iloc[y_means==4,0],X.iloc[y_means==4,1],s=100,c="red",label="Cluster 5")
pse.title("Customer Segmentation")
pse.xlabel("Annual Income")
pse.ylabel("Spending Score")
pse.legend()
pse.show()
```
