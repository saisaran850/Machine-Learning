```python
PLW.figure(1,figsize=(15,5))
n=0;
for x in ["MargAdh","NormNucl","Mit"]:
    n += 1
    PLW.subplot(1,3,n)
    PLW.subplots_adjust(hspace=0.5,wspace=0.5)
    SRE.stripplot(df[x])
    PLW.title("Displot of {}".format(x))
PLW.show()
```

```python
PLW.figure(figsize=(15,5))
SRE.countplot(y="SingEpiSize",data=df)
PLW.show()
```

```python
# Defining Subset
Subset = df.drop(["MargAdh","SingEpiSize","Mit"],axis=1)
```

```python
# Dendrogram
PLW.figure(figsize=(10,7))
PLW.title("Dendrograms")
dendrogram = SCH.dendrogram(SCH.linkage(Subset,method="ward"))
```

```python
# Predict Subset
h_cluster = AgglomerativeClustering(n_clusters=3,affinity="euclidean")
h_cluster.fit_predict(Subset)
```
