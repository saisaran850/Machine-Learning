```python
# Defining X
import pandas as PT
df = PT.read_csv("D:\Dataset\Apple Dataset.csv")
df
X = df.iloc[:,[1,2]].values
```

```python
# Importing DBSCAN
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=3,min_samples=4)
```

```python
# Fitting the model
model = dbscan.fit(X)
labels = model.labels_
```

```python
labels
```

```python
# Importing metrics
from sklearn import metrics
import numpy as NR
sample_cores=NR.zeros_like(labels,dtype=bool)
```

```python
sample_cores
```

```python
sample_cores[dbscan.core_sample_indices_]=True
sample_cores
```
